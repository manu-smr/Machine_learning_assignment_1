```python
def median(l):
    div=str(len(l)/2)
    decimal=int(div[div.find('.')+1])
    if(decimal>0):
        med=l[int(div[:div.find('.')])]
    else:
        med=(l[(int(div[:div.find('.')])-1)]+l[(int(div[:div.find('.')]))])/2
    return med

def average(l):
    total=0
    for i in l:
        total=total+i
    return total/len(l)
```
